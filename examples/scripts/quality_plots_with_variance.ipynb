```python
# Calculate 95% Confidence Intervals for the mean across runs (n=5)
print("Calculating 95% confidence intervals for the mean across runs...")

def calculate_95_ci_halfwidth(row):
    """
    Calculate 95% CI half-width for the mean across n=5 runs.
    Uses t-distribution: halfwidth = t_{0.975,4} * s / sqrt(5)
    where t_{0.975,4} = 2.776 for 95% CI with df=4
    """
    if 'results' in row and row['results'] is not None and len(row['results']) > 1:
        results = np.array(row['results'])
        n = len(results)
        
        # Sample standard deviation (Bessel's correction)
        s = np.std(results, ddof=1)
        
        # For n=5, df=4, t_{0.975,4} = 2.776
        t_critical = 2.776
        
        # 95% CI half-width
        halfwidth = t_critical * s / np.sqrt(n)
        return halfwidth
    else:
        return 0.0

# Calculate confidence interval half-widths
df['ci_halfwidth'] = df.apply(calculate_95_ci_halfwidth, axis=1)

# Also recalculate sample standard deviation for reference
df['std'] = df.apply(lambda row: np.std(row['results'], ddof=1) if 'results' in row and row['results'] is not None and len(row['results']) > 1 else row.get('std', 0), axis=1)

# Show comparison of SD vs 95% CI half-width for a few examples
print("\nComparison of Standard Deviation vs 95% CI half-width (first 5 rows):")
print("Note: CI half-width represents uncertainty in the MEAN estimate")
for i in range(min(5, len(df))):
    row = df.iloc[i]
    sd = row['std']
    ci_hw = row['ci_halfwidth']
    ratio = ci_hw / sd if sd > 0 else 0
    print(f"{row['experiment'][:30]}: SD = {sd:.4f}, CI half-width = {ci_hw:.4f}, Ratio = {ratio:.2f}")

print(f"\n95% Confidence intervals calculated using t-distribution (t_0.975,4 = 2.776)")
print("Error bars will now represent uncertainty in the mean estimate, not run-to-run variability")

# Helper function to extract data with 95% CIs for plotting
def get_data_with_ci(data_df, filters):
    """
    Extract mean and 95% CI half-width from dataframe with given filters
    Returns values scaled to percentage (x100)
    """
    filtered_data = data_df
    for key, value in filters.items():
        filtered_data = filtered_data[filtered_data[key] == value]
    
    if len(filtered_data) > 0:
        row = filtered_data.iloc[0]
        return row['mean'] * 100, row['ci_halfwidth'] * 100
    else:
        return 0, 0
```

```python
# Plot 1 (UPDATED): Final performance on QuaLITY with 95% confidence intervals
plt.figure(figsize=(14, 8))

# Get final performance data (default dataset type, chunk 2 or full)
final_data = []

# Baseline
baseline_mean, baseline_ci = get_data_with_ci(df, {
    'experiment': 'baseline', 'dataset_type': 'default'
})
final_data.append({
    'name': 'Baseline',
    'mean': baseline_mean,
    'ci_halfwidth': baseline_ci,
    'method': 'baseline'
})

# OSFT experiments (both learning rates)
for lr in ['2e-5', '5e-6']:
    # Chunked decreasing (chunk 2)
    dec_mean, dec_ci = get_data_with_ci(df, {
        'lr': lr, 'experiment': 'osft-chunked-decreasing-chunk2', 'dataset_type': 'default'
    })
    final_data.append({
        'name': f'OSFT Decreasing RR\n(Chunk 3, {lr})',
        'mean': dec_mean,
        'ci_halfwidth': dec_ci,
        'method': 'osft'
    })
    
    # Chunked constant (chunk 2)
    const_mean, const_ci = get_data_with_ci(df, {
        'lr': lr, 'experiment': 'osft-chunked-chunk2', 'dataset_type': 'default'
    })
    final_data.append({
        'name': f'OSFT Constant RR\n(Chunk 3, {lr})',
        'mean': const_mean,
        'ci_halfwidth': const_ci,
        'method': 'osft'
    })
    
    # Full dataset
    full_mean, full_ci = get_data_with_ci(df, {
        'lr': lr, 'experiment': 'osft-full', 'dataset_type': 'default'
    })
    final_data.append({
        'name': f'OSFT Full Dataset\n({lr})',
        'mean': full_mean,
        'ci_halfwidth': full_ci,
        'method': 'osft'
    })

# SFT experiments
# SFT chunked (chunk 2)
sft_chunked_mean, sft_chunked_ci = get_data_with_ci(df, {
    'lr': '2e-5', 'experiment': 'sft-chunked-chunk2', 'dataset_type': 'default'
})
final_data.append({
    'name': 'SFT Chunked\n(Chunk 3)',
    'mean': sft_chunked_mean,
    'ci_halfwidth': sft_chunked_ci,
    'method': 'sft'
})

# SFT full
sft_full_mean, sft_full_ci = get_data_with_ci(df, {
    'lr': '2e-5', 'experiment': 'sft-full', 'dataset_type': 'default'
})
final_data.append({
    'name': 'SFT Full Dataset',
    'mean': sft_full_mean,
    'ci_halfwidth': sft_full_ci,
    'method': 'sft'
})

# Create bar plot with 95% CIs
x = np.arange(len(final_data))
means = [d['mean'] for d in final_data]
ci_halfwidths = [d['ci_halfwidth'] for d in final_data]
methods = [d['method'] for d in final_data]

# Color mapping
colors = []
for m in methods:
    if m == 'baseline':
        colors.append('#333333')
    elif m == 'osft':
        colors.append('#1f77b4')
    else:  # sft
        colors.append('#ff7f0e')

bars = plt.bar(x, means, yerr=ci_halfwidths, capsize=5, color=colors, alpha=0.8, edgecolor='black')

# Add value labels
for i, (bar, mean, ci_hw) in enumerate(zip(bars, means, ci_halfwidths)):
    # Position labels slightly above error bars
    label_y = bar.get_height() + ci_hw + 0.3
    # Check if label would go above plot area and adjust if needed
    if label_y > 54.5:
        label_y = bar.get_height() - ci_hw - 0.5
        va = 'top'
    else:
        va = 'bottom'
    plt.text(bar.get_x() + bar.get_width()/2, label_y, 
             f'{mean:.1f}%', ha='center', va=va, fontsize=10)

plt.xlabel('Method', fontsize=12)
plt.ylabel('QuaLITY Accuracy (%)', fontsize=12)
plt.title('Final Performance on QuaLITY Dataset (95% CIs, n=5 runs)\nComparing Learning Rates 2e-5 vs 5e-6', fontsize=14)
plt.xticks(x, [d['name'] for d in final_data], rotation=45, ha='right')
plt.ylim(38, 55)  # Focused range to better show differences
plt.grid(axis='y', alpha=0.3)

# Add more y-axis ticks for better readability
plt.yticks(np.arange(38, 56, 2))

# Add legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='#333333', edgecolor='black', label='Baseline'),
    Patch(facecolor='#1f77b4', edgecolor='black', label='OSFT'),
    Patch(facecolor='#ff7f0e', edgecolor='black', label='SFT')
]
plt.legend(handles=legend_elements, loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Plot 7 (FINAL): Pareto Frontier with Percentage-based Backward Transfer
fig, ax = plt.subplots(figsize=(12, 10))

def calculate_backward_transfer_percent(lr, method_base, experiment_name):
    """
    Calculate backward transfer metric as percentage:
    For each chunk: (initial - final) / initial * 100
    Then average across chunks
    Range: [0, 100%] where 0% = no forgetting, 100% = complete forgetting
    """
    relative_forgetting_scores = []
    
    for chunk in [0, 1]:  # chunks 1 and 2 (0-indexed)
        # Initial score (single-chunk evaluation) - convert to raw proportion
        initial_exp = f'{method_base}-chunk{chunk}' if method_base != 'sft-chunked' else f'sft-chunked-chunk{chunk}'
        initial_mean, _ = get_data_with_ci(df_with_forgetting, {
            'lr': lr, 'dataset_type': 'single-chunk', 'experiment': initial_exp
        })
        initial_mean = initial_mean / 100  # Convert back to proportion
        
        # Final score (forgetting evaluation) - convert to raw proportion
        final_exp = f'{experiment_name}-final-on-chunk{chunk}'
        final_mean, _ = get_data_with_ci(df_with_forgetting, {
            'lr': lr, 'dataset_type': 'forgetting', 'experiment': final_exp
        })
        final_mean = final_mean / 100  # Convert back to proportion
        
        if initial_mean > 0 and final_mean > 0:
            # Relative forgetting = (initial - final) / initial
            # This gives us the fraction of performance lost
            relative_forgetting = (initial_mean - final_mean) / initial_mean
            relative_forgetting_scores.append(relative_forgetting)
    
    # Average relative forgetting across chunks, convert to percentage
    if len(relative_forgetting_scores) > 0:
        return np.mean(relative_forgetting_scores) * 100  # Convert to percentage
    else:
        return 0

# Calculate metrics for each method
methods_for_pareto_final = [
    ('sft-chunked', 'sft-chunked', 'SFT Chunked', '#ff7f0e', '2e-5', '^'),
    ('osft-chunked', 'osft-chunked', 'OSFT Constant RR (2e-5)', '#2ca02c', '2e-5', 's'),
    ('osft-chunked-decreasing', 'osft-chunked-decreasing', 'OSFT Decreasing RR (2e-5)', '#1f77b4', '2e-5', 'o'),
    ('osft-chunked', 'osft-chunked', 'OSFT Constant RR (5e-6)', '#8c564b', '5e-6', 's'),
    ('osft-chunked-decreasing', 'osft-chunked-decreasing', 'OSFT Decreasing RR (5e-6)', '#9467bd', '5e-6', 'o'),
]

pareto_data_final = []

for method_base, experiment_name, label, color, lr, marker in methods_for_pareto_final:
    backward_transfer_pct = calculate_backward_transfer_percent(lr, method_base, experiment_name)
    quality_gain, quality_gain_ci = calculate_quality_gain_with_ci_updated(lr, method_base)
    
    pareto_data_final.append({
        'label': label,
        'backward_transfer': backward_transfer_pct,
        'quality_gain': quality_gain,
        'quality_gain_ci': quality_gain_ci,
        'color': color,
        'marker': marker,
        'lr': lr
    })
    
    print(f"{label}: Quality Gain = {quality_gain:.2f}% ({quality_gain_ci:.2f}%), Backward Transfer = {backward_transfer_pct:.2f}%")

# Create the pareto plot with error bars
for data in pareto_data_final:
    ax.errorbar(data['backward_transfer'], data['quality_gain'], 
                yerr=data['quality_gain_ci'],
                marker=data['marker'], markersize=12, 
                color=data['color'], alpha=0.8, 
                markeredgecolor='white', markeredgewidth=2,
                capsize=3, capthick=1.5,
                label=data['label'], linestyle='None')

# Add quadrant lines
ax.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
ax.axvline(x=0, color='gray', linestyle='--', alpha=0.5)

# Add annotations for quadrants
ax.text(0.02, 0.95, 'High Quality Gain\nNo Forgetting\n(BEST)', 
        transform=ax.transAxes, ha='left', va='top',
        bbox=dict(boxstyle='round,pad=0.5', facecolor='lightgreen', alpha=0.7))

ax.text(0.98, 0.05, 'Low Quality Gain\nHigh Forgetting\n(WORST)', 
        transform=ax.transAxes, ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', facecolor='lightcoral', alpha=0.7))

# Customize plot
ax.set_xlabel('Backward Transfer (Forgetting) %\n Less Forgetting | More Forgetting ', fontsize=12)
ax.set_ylabel('Quality Gain over Baseline (%)\n Lower Gain | Higher Gain ', fontsize=12)
ax.set_title('Pareto Frontier: Quality Improvement vs Knowledge Retention\nComparing Training Methods (95% CIs, n=5 runs)', fontsize=14)
ax.grid(True, alpha=0.3)
ax.legend(loc='upper right', fontsize=11, labelspacing=0.6)

plt.tight_layout()
plt.show()
```

## Summary (Updated)

This notebook has been updated to use **95% confidence intervals** instead of standard deviation bars. This provides a more appropriate statistical representation for comparing model performance.

### Statistical Approach:
- **Error bars represent**: 95% confidence intervals for the mean performance estimate
- **Formula**: `halfwidth = t_{0.975,4}  s / 5 = 2.776  s / 5`
- **Interpretation**: We can be 95% confident the true mean performance lies within this interval
- **Why this is better**: CIs tell us how precisely we know the mean, not just run-to-run variability

### Plots Created:
1. **Final Performance Bar Chart** - 95% CIs for final accuracy comparison
2. **Temporal Performance Plot** - Performance evolution with confidence bands  
3. **Domain-Specific Evaluation** - Split by learning rate with 95% CIs
4. **Progressive Evaluation Bar Chart** - Knowledge retention with 95% CIs
5. **Progressive Evaluation Temporal** - Retention trends with confidence bands
6. **Forgetting Analysis (2e-5)** - Original vs Final performance with 95% CIs
7. **Forgetting Analysis (5e-6)** - OSFT 5e-6 vs SFT baseline with 95% CIs
8. **Pareto Frontier Plot** - Quality improvement vs forgetting trade-off analysis

### Key Insights for Forgetting Analysis:
- **Light bars**: Original performance immediately after training on that chunk
- **Dark bars**: Final performance after completing all training (measures forgetting)
- **95% CIs**: Show uncertainty in our mean estimates, allowing proper statistical comparison
- **Interpretation**: Non-overlapping CIs suggest statistically significant differences

### Pareto Frontier Analysis:
- **X-axis (Backward Transfer)**: Average forgetting across chunks 1 and 2
  - Formula: `((initial_chunk1 - final_chunk1) + (initial_chunk2 - final_chunk2)) / 2`
  - Positive values = more forgetting, Negative values = improvement over time
- **Y-axis (Quality Gain)**: Performance improvement over baseline on final evaluation
- **Ideal methods**: Upper-left quadrant (high quality gain, low forgetting)
- **Poor methods**: Lower-right quadrant (low quality gain, high forgetting)

This statistical rigor allows for more confident conclusions about which training methods truly differ in their ability to retain knowledge from earlier chunks while achieving strong final performance.


```python
# Plot 6a: Forgetting Analysis - 2e-5 Learning Rate (OSFT vs SFT)
fig, ax = plt.subplots(figsize=(14, 8))

# Define the evaluation chunks we have data for
eval_chunks = [0, 1]  # chunk 0 and chunk 1
chunk_labels = ['Chunk 1', 'Chunk 2']

# Get original and final performance for each method
def get_forgetting_data(lr, method_base, experiment_name):
    """Extract original and final performance for a given method"""
    original_scores = []
    original_stds = []
    final_scores = []
    final_stds = []
    
    for chunk in eval_chunks:
        # Original score (single-chunk evaluation)
        original_exp = f'{method_base}-chunk{chunk}' if method_base != 'sft-chunked' else f'sft-chunked-chunk{chunk}'
        original_data = df_with_forgetting[
            (df_with_forgetting['lr'] == lr) & 
            (df_with_forgetting['dataset_type'] == 'single-chunk') & 
            (df_with_forgetting['experiment'] == original_exp)
        ]
        
        if len(original_data) > 0:
            original_scores.append(original_data.iloc[0]['mean'] * 100)
            original_stds.append(original_data.iloc[0]['std'] * 100)
        else:
            original_scores.append(0)
            original_stds.append(0)
            
        # Final score (forgetting evaluation)
        final_exp = f'{experiment_name}-final-on-chunk{chunk}'
        final_data = df_with_forgetting[
            (df_with_forgetting['lr'] == lr) & 
            (df_with_forgetting['dataset_type'] == 'forgetting') & 
            (df_with_forgetting['experiment'] == final_exp)
        ]
        
        if len(final_data) > 0:
            final_scores.append(final_data.iloc[0]['mean'] * 100)
            final_stds.append(final_data.iloc[0]['std'] * 100)
        else:
            final_scores.append(0)
            final_stds.append(0)
    
    return original_scores, original_stds, final_scores, final_stds

# Methods for 2e-5 plot
methods_2e5 = [
    ('sft-chunked', 'sft-chunked', 'SFT', '#ff7f0e'),
    ('osft-chunked', 'osft-chunked', 'OSFT Constant RR', '#2ca02c'),
    ('osft-chunked-decreasing', 'osft-chunked-decreasing', 'OSFT Decreasing RR', '#1f77b4'),
]

# Set up bar positions
bar_width = 0.12
n_methods = len(methods_2e5)
n_scores = 2  # original and final
x = np.arange(len(eval_chunks))

# Plot bars for each method
bar_index = 0
for method_base, experiment_name, label, color in methods_2e5:
    orig_scores, orig_stds, final_scores, final_stds = get_forgetting_data('2e-5', method_base, experiment_name)
    
    # Original scores (lighter shade)
    bars_orig = ax.bar(x + bar_index * bar_width - (n_methods*n_scores-1)*bar_width/2, 
                       orig_scores, bar_width, yerr=orig_stds, capsize=3,
                       label=f'{label} (Original)', color=color, alpha=0.6, edgecolor='black')
    bar_index += 1
    
    # Final scores (darker shade)
    bars_final = ax.bar(x + bar_index * bar_width - (n_methods*n_scores-1)*bar_width/2, 
                        final_scores, bar_width, yerr=final_stds, capsize=3,
                        label=f'{label} (Final)', color=color, alpha=1.0, edgecolor='black')
    bar_index += 1

ax.set_xlabel('Evaluation Chunk', fontsize=12)
ax.set_ylabel('QuaLITY Accuracy (%)', fontsize=12)
ax.set_title('Forgetting Analysis: Original vs Final Performance (Learning Rate: 2e-5)', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(chunk_labels)
ax.legend(loc='upper right', ncol=2)
ax.grid(axis='y', alpha=0.3)
ax.set_ylim(45, 58)

plt.tight_layout()
plt.show()
```

```python
# Plot 6b: Forgetting Analysis - 5e-6 Learning Rate (OSFT vs SFT Baselines)
fig, ax = plt.subplots(figsize=(14, 8))

# Methods for 5e-6 plot (include SFT baselines from 2e-5 for comparison)
methods_5e6 = [
    ('sft-chunked', 'sft-chunked', 'SFT (2e-5)', '#ff7f0e', '2e-5'),  # SFT baseline
    ('osft-chunked', 'osft-chunked', 'OSFT Constant RR (5e-6)', '#8c564b', '5e-6'),
    ('osft-chunked-decreasing', 'osft-chunked-decreasing', 'OSFT Decreasing RR (5e-6)', '#9467bd', '5e-6'),
]

# Plot bars for each method
bar_index = 0
for method_base, experiment_name, label, color, lr in methods_5e6:
    orig_scores, orig_stds, final_scores, final_stds = get_forgetting_data(lr, method_base, experiment_name)
    
    # Original scores (lighter shade)
    bars_orig = ax.bar(x + bar_index * bar_width - (n_methods*n_scores-1)*bar_width/2, 
                       orig_scores, bar_width, yerr=orig_stds, capsize=3,
                       label=f'{label} (Original)', color=color, alpha=0.6, edgecolor='black')
    bar_index += 1
    
    # Final scores (darker shade)
    bars_final = ax.bar(x + bar_index * bar_width - (n_methods*n_scores-1)*bar_width/2, 
                        final_scores, bar_width, yerr=final_stds, capsize=3,
                        label=f'{label} (Final)', color=color, alpha=1.0, edgecolor='black')
    bar_index += 1

ax.set_xlabel('Evaluation Chunk', fontsize=12)
ax.set_ylabel('QuaLITY Accuracy (%)', fontsize=12)
ax.set_title('Forgetting Analysis: Original vs Final Performance (OSFT 5e-6 vs SFT Baseline)', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(chunk_labels)
ax.legend(loc='upper right', ncol=2)
ax.grid(axis='y', alpha=0.3)
ax.set_ylim(45, 58)

plt.tight_layout()
plt.show()
```
